### Multiclass Woe Pipeline

```python
# multiclass data set
import pandas as pd

data_source = "/home/fra/DataMart/datacentre/opendata/UCI/"
train_file = "iris.csv"

# load data
df = pd.read_csv(data_source + train_file)
df.head()
```

```python
def f(X, t25, t50, t75):
    if X<t25:
        return 'low'
    elif X<t50:
        return 'mid'
    elif X<t75:
        return 'long'
    else:
        return 'very long'
```

```python
df['petal_width_category'] = df['petal_width'].apply(f, args=(0.3, 1.3, 1.8))
df = df.drop(['petal_width'], axis=1)
df.head()
```
