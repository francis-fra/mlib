```python
def f(X, t25, t50, t75):
    if X<t25:
        return 'low'
    elif X<t50:
        return 'mid'
    elif X<t75:
        return 'long'
    else:
        return 'very long'
```

```python
df['petal_width_category'] = df['petal_width'].apply(f, args=(0.3, 1.3, 1.8))
```

### Modelling

```python
df.head()
```

```python
import pipeline as pl
import utility as ut
```

```python
(df_train, y_train, df_test, y_test) = ut.df_train_test_split(df, "species", test_size=0.6, dfout=True)
```

```python
df_train['species'] = y_train
df_test['species'] = y_test
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
stdpl = pl.StandardPipeline("SPECIES", [])
stdpl.make_pipeline(ntransformer=StandardScaler, ctransformer=OrdinalEncoder)
```

```python
(X_train, y_train) = stdpl.fit_transform(df_train)
(X_test, y_test) = stdpl.transform(df_test)
```

```python
X_train.shape, X_test.shape
```

```python
log_reg.fit(X_train, y_train)
```

```python
y_test == log_reg.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, log_reg.predict_proba(X_test), multi_class="ovr") 
```
